Task 1:
Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce() 


```python
def my_reduce(function, sequence):
    it = iter(sequence)
    value = next(it)
    for element in it:
        value = function(value, element)

    return value

print("Total sum is ", my_reduce(lambda x, y: x+y, [1, 2, 3, 4, 5]))
print("Biggest number is ", my_reduce(lambda x, y: x if x > y else y, [1, 2, 3, 4, 5]))
print("Smallest number is ", my_reduce(lambda x, y: x if x < y else y, [1, 2, 3, 4, 5]))
print("Multiply of sequence is", my_reduce(lambda x, y: x*y, [1, 2, 3, 4, 5]))
```

Task 2: Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter() 

```python
l = [ 1, 2, 3, 4, 5,6]

def my_filter(function, sequence):
    it = iter(sequence)
    for element in it:
        value = function(element)
        if value == True:
          yield element

print("odd numbers filtering from list ", list(my_filter(lambda x: x%2 !=0 , l)))
print("even numbers filtering from list ", list(my_filter(lambda x: x%2 ==0 , l)))
```

Task 4: Implement a function longestWord() that takes a list of words and returns the longest one. 

```python
import functools
def longestWord(list_of_words):
  return functools.reduce(lambda x,y: x if len(x) > len(y) else y, list_of_words)

list_of_words = ['hey', 'this', 'is', 'a', 'toughest' , 'assignment']
print("longestword is :",longestWord(list_of_words))
```

Write a Python Program(with class concepts) to find the area of the triangle using the below formula. 
 
area = (s*(s-a)*(s-b)*(s-c)) ** 0.5 
 
Function to take the length of the sides of triangle from user should be defined in the parent class and function to calculate the area should be defined in subclass.


```python
class Triange:
    def __init__(self, a,b,c):
        self.a = a
        self.b = b
        self.c = c
class Area: 
    def __init__(self,Triange):
        self.Triange = Triange
        
    def side_length(self):
        return (self.Triange.a+self.Triange.b+self.Triange.c)/2
    
    def cal_area(self,side):
       return (side * (side - self.Triange.a) * (side - self.Triange.b) * (side - self.Triange.c))  ** 0.5

side1 = int(input("Enter side1 length "))
side2 = int(input("Enter side1 length "))
side3 = int(input("Enter side1 length "))

triange = Triange(side1,side2,side3)
area = Area(triange)
print('The area of the triangle is %0.2f' %area.cal_area(area.side_length()))
```

Task 6: 
Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n. 

```python
def  filter_long_words(list_of_words,input_length):
  return filter(lambda x: len(x) > input_length, list_of_words)

input_length = 4
list_of_words = ['hey', 'this', 'is', 'a', 'toughest' , 'assignment' , 'sudhansu', 'please', 'give', 'more' , 'hints']
print("long words then given length are:",  list(filter_long_words(list_of_words,input_length)))
```

Task 7:
    Write a Python program using function concept that maps  list of words into a list of integers representing the lengths of the corresponding words. 
 Hint: If a list [ ab,cde,erty] is passed on to the python function output should come as [2,3,4] 
 Here 2,3 and 4 are the lengths of the words in the list.

```python
def  length_of_each_word(list_of_words):
  return map(lambda x: len(x), list_of_words)

list_of_words = ['ab','cde','erty'] 
length_of_words = list(length_of_each_word(list_of_words))
print('length of each word in the list:', length_of_words)
```

Task 8: Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise. 

```python
def  find_input_is_vowel(input_character,list_of_vowels):
  return True if list_of_vowels.count(input_character) > 0 else False

list_of_vowels = ['a','e','i','o','u','A','E','I','O','U']
input_character = input("Enter only one character ")
while(1):
  if (len(input_character) != 1):
    print('Please enter only one character ')
    input_character = input("Enter only one character ")
  else:
    break
print("given input is vowel ", find_input_is_vowel(input_character,list_of_vowels))
```
